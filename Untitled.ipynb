```python
#boundries across seasons
data=pd.DataFrame()

sixes=df5.groupby('season')['batsman_runs'].apply(lambda x:(x==6).sum())
fours=df5.groupby('season')['batsman_runs'].apply(lambda x:(x==4).sum())
boundries=pd.DataFrame({'season':sixes.index,'6':sixes.values})
boundries['4']=fours.values
boundries
boundries[['6','4']].plot(marker='o')
fig=plt.gcf()
fig.set_size_inches(10,6)
plt.show()




```

```python
runs_season_wise=df5.groupby(['season','batting_team'])['total_runs'].sum()

runs_season_wise.index.get_level_values('season')
runs_season_wise.index.get_level_values('batting_team')
trunSWise=pd.DataFrame({'season':runs_season_wise.index.get_level_values('season'),'batting_team':runs_season_wise.index.get_level_values('batting_team'),'runs':runs_season_wise.values})

p = trunSWise.pivot(index='season', columns='batting_team')
team_runs_sea=pd.DataFrame({'CSK':p.values[:,0]})
#team_runs_sea['DC']=p.values[:,1]
team_runs_sea['DD']=p.values[:,2]
#team_runs_sea['GL']=p.values[:,3]
team_runs_sea['KKR']=p.values[:,4]
#team_runs_sea['KTK']=p.values[:,5]
team_runs_sea['KXI']=p.values[:,6]
team_runs_sea['MI']=p.values[:,7]
#team_runs_sea['PW']=p.values[:,8]
team_runs_sea['RCB']=p.values[:,9]
#team_runs_sea['RPS']=p.values[:,10]
team_runs_sea['RR']=p.values[:,11]
team_runs_sea['SRH']=p.values[:,12]
team_runs_sea=team_runs_sea.fillna(0)
team_runs_sea



```

```python
team_runs_sea.plot(marker='o',color=["b", "r", "#Ffb6b2", "g",'brown','y','#6666ff','black','#FFA500'])
fig=plt.gcf()
fig.set_size_inches(12,8)
plt.show()
```
